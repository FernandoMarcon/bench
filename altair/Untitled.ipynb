```python
import pandas as pd
import altair as alt
```

```python
alt.renderers.enable('mimetype')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/onlyphantom/miband/main/data/run_1km.csv',parse_dates=['startTime'])
```

```python
df['day_of_week'] = df['startTime'].dt.day_name()
```

```python
alt.Chart(df).mark_point().encode(
    # x,y,color,shape,size
    # x='seconds_per_km:Q',
    # y="day_of_week:N"
    alt.X('day_of_week'),
    alt.Y('seconds_per_km')
).interactive()
```

```python
df.dtypes
```

Q -> Quantitative (numerical data)
N -> Nominal (unordered, categorical)
O -> Ordinal (rank-ordered)
T -> Temporal (date/time)
